```python
import pandas as pd

# Load the datasets
#incart_df = pd.read_csv('/mnt/data/INCART 2-lead Arrhythmia Database.csv')
mit_bih_df = pd.read_csv('/content/MIT-BIH Arrhythmia Database.csv')
mit_bih_sup_df = pd.read_csv('/content/MIT-BIH Supraventricular Arrhythmia Database.csv')
```

```python
import matplotlib.pyplot as plt

# Assuming the 'record' column contains time-related data and the other columns are ECG signal features

# Extracting a subset of the MIT-BIH Arrhythmia dataset for visualization
mit_bih_subset = mit_bih_df.head(1000)  # Taking first 1000 records for visualization

# Plotting the ECG signals from lead 0 and lead 1 over time
plt.figure(figsize=(12, 6))

# Plotting lead 0's ECG signal (e.g., '0_rPeak')
plt.subplot(2, 1, 1)
plt.plot(mit_bih_subset.index, mit_bih_subset['0_rPeak'], label='Lead 0 - rPeak')
plt.title('MIT-BIH Arrhythmia Data - Lead 0 rPeak')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Plotting lead 1's ECG signal (e.g., '1_rPeak')
plt.subplot(2, 1, 2)
plt.plot(mit_bih_subset.index, mit_bih_subset['1_rPeak'], label='Lead 1 - rPeak', color='orange')
plt.title('MIT-BIH Arrhythmia Data - Lead 1 rPeak')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
